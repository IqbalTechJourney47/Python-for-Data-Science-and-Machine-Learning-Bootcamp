# Indexing and slicing of array

```python
arr = np.arange(0,11)
```

```python
arr
```

```python
arr[8]
```

```python
arr[1:5]
```

```python
arr[0:5]
```

```python
arr[:6]
```

```python
arr[0:6]
```

```python
arr[5:]
```

```python
# broadcasting value equal to 100
# 0-4, 1st five elements are 100 value

arr[0:5] = 100
```

```python
arr
```

```python
arr = np.arange(0,11)
```

```python
arr
```

```python
slice_of_arr = arr[0:6]
```

```python
slice_of_arr
```

```python
# broadcasting

slice_of_arr[:]
```

```python
slice_of_arr[:] = 99
```

```python
slice_of_arr
```

```python
arr
```

```python
# If you actually grab a slice of array and set it as a variable without explicitly saying that you want to
# copy of the array, you are just viewing a link to the original array and that changes you do will actually
# affect that original array.

arr_copy = arr.copy()
```

```python
arr
```

```python
arr_copy
```

```python
arr_copy[:] = 100
```

```python
arr_copy
```

```python
arr
```

# 2-D array

```python
arr_2d = np.array([[5,10,15],[20,25,30],[35,40,45]])
```

```python
arr_2d
```

```python
# 1st row and 1st column

arr_2d[0][0]
```

```python
# entire row of index 0

arr_2d[0]
```

```python
# index 1 of the row 1
# from the row 1, index 1

arr_2d[1][1]
```

```python
# from the row 2, index 1

arr_2d[2][1]
```

```python
arr_2d[2,1]
```

```python
arr_2d[1,2]
```

```python
arr_2d[1][2]
```

```python
arr_2d = np.array([[5,10,15],[20,25,30],[35,40,45]])
arr_2d
```

```python
# grab everything upto column 2 and then grab from column 2 onwards
# grab everything but not including rows 0 and 1 as the colon
# but then don't include two, we are saying slice it too

# grab everything from column 1 all the way to end
# drop column 0

# :2 - rows
# 1: - columns

arr_2d[:2,1:]
```

```python
arr_2d[:2]
```
