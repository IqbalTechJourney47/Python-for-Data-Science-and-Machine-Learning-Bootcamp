# K Means Clustering (Unsupervised)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
# centers=4 , blobs=4

data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
# K Means - unsupervised algorithm
# This algorithm randomly assigns each observation to a cluster and then finds a centroid of each cluster.
# Then the algorithm will iterate through two steps
# It re-assigned data points to the cluster whose centroid is closest.
# It calculates the new centre for each cluster
# and then it just repeats that over and over again until the cluster variation can't be reduced any further.
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
# Predicted labels for data

kmeans.labels_
```

```python
# If you are working of real data and you didn't have the labels, you would be done at this stage
# you would not be able to actually compare anything to the real values
# But since we actually do know the labels from data ones
# K means clustering is unsupervised, so you wouldn't use it to actually predict labels,
# you would use it to try to find labels in your data
```

```python
# sharey = share same axis
# data[0][:,0] = all the rows in the first column
# data[0][:,1] = all the rows in the second column

# subplots(1,2) = subplot of 1 row by 2 columns
fig , (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

# Plot original data and color that data based off of what my algorithm for the label should look like
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

# Original data
# Plot original data but it's going to colour them by they're actually factually correct blobs assignments
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
# How it differs from the original
# In K Means
# In this noisy section, where our algorithm wasn't able to really detect which group belong to which point
# or which point belong to which group

# In the true dataset, there's some noise but essentially Kmeans just decided to deal with that noise and assign them
# all the same centroid cluster there

# These colors are just meaningless and they only really referenced within each other own plots

# That doesn't mean that yellow is the same as yellow or blue is the same as blue as the original
# It's just the way of coloring stuff
```
