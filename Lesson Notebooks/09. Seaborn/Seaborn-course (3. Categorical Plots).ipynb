```python
import seaborn as sns
```

```python
%matplotlib inline
```

# Categorical Plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Bar Plot

```python
# Bar plot is a general plot that allows you to aggregate categorical data based off of some function
# and by default that's the mean

# x = Set the x axis to a categorical column
# Y = We can choose a column that's numeric for instance

# It is showing the average or mean of the total_bill per categorical column value
# Average total_bill for males is going to be slightly higher than the average total_bill for females

sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np
```

```python
# Bar plot is a general plot that allows you to aggregate categorical data based off of some function
# and by default that's the mean

# x = Set the x axis to a categorical column
# Y = We can choose a column that's numeric for instance

# It is showing the average or mean of the total_bill per categorical column value
# Average total_bill for males is going to be slightly higher than the average total_bill for females

# estimator(x, y, hue, data)
# estimator = Statistical function to estimate within each categorical bin

# standard deviation of total_bill column per category

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

# Count Plot

```python
# Count plot is same as bar plot except the estimator is explicitly counting the number of occurences

# It gives the count of the tips column of gender.

sns.countplot(x='sex', data=tips)
```

# Box plot and voilin plot

```python
# These type of plots are used to that distribution of categorical data,
# a box plot is also sometimes known as box and whisker plot

# Box plot shows the distribution of quantitative data in a way that hopefully facilitates comparisons between the variables

# x= a categorical column day
# y= a numerical column

# It gives the distribution of total_bill per day

sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# Violin plot plays a very similar role to a box plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

# Strip plot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

# Swarm plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

# Factor plot

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```
